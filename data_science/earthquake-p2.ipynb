```python
import pandas as pd #library for data structures and analysis
import numpy as np #multi-dimensional container and N-Dimensional array object
```

```python
train_df = pd.read_csv('database.csv')
train_df.drop(columns=['Depth Error', 'Depth Seismic Stations', 'Magnitude Error', 'Magnitude Seismic Stations', 'Azimuthal Gap', 'Horizontal Distance', 'Horizontal Error', 'Root Mean Square'], inplace=True)
train_df.head()
```

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt 

from sklearn.neighbors import KernelDensity
from scipy.stats import gaussian_kde
from statsmodels.nonparametric.kde import KDEUnivariate
from statsmodels.nonparametric.kernel_density import KDEMultivariate


%matplotlib inline

sns.set(style="ticks")

sns.pairplot(train_df, hue = 'ID')
plt.show()
```

```python
test_df = pd.read_csv('test.csv')
test_df.head()
```

```python
# One-hot encode the data using pandas get_dummies
# Convert string variables to integer so we can load it to ML Algorithms
train_df = pd.get_dummies(train_df)
# Display the first 5 rows of the last 12 columns
train_df.iloc[:,5:].head(5)
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(train_df['rate_of_penetration'])
# Remove the labels from the features
# axis 1 refers to the columns
train_df= train_df.drop('rate_of_penetration', axis = 1)
# Saving feature names for later use
train_df_list = list(train_df.columns)
# Convert to numpy array
train_df = np.array(train_df)
```

```python
#the split
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
#orig : train_features, test_features, train_labels, test_labels = train_test_split(train_df, labels, test_size = 0.2, random_state = 42)
#Set test_size to None
train_features, test_features, train_labels, test_labels = train_test_split(train_df, labels, test_size = None, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
#TRAIN the MODEL
# Import the model we are using
#Random forest is a Supervised Learning algorithm which uses ensemble learning method for classification and regression
#Aggregate many decision trees
#Bagging technique; runs decision trees in parallel without interaction with other trees 
from sklearn.ensemble import RandomForestRegressor

# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)

# Train the model on training data
rf.fit(train_features, train_labels);

#print list(rf);
```

# Make Predictions on the Test Set

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2))
print ('####### Predictions #######');
#print list(predictions);

#stack predictions vertically and create predictions output file
foutl = np.vstack((predictions))
print foutl
np.savetxt('foutpreds.txt', foutl)
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
