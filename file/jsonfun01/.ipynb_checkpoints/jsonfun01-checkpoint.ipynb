### string to json

```python
import json 
  
# Sample json string 
#============================
employee ='{"id":"09", "name": "Benjamin", "department":"IT"}'
  

# Convert string to Python dict 
#============================
employee_dict = json.loads(employee) 

# Output
#============================
print(employee_dict) 
print(employee_dict['name']) 
```

### Read from json file

```python
import json 
   
# Opening JSON file 
f = open('apperian-devicedetails-20200717.json',) 
   
# returns JSON object as  
# a dictionary 
data = json.load(f) 
   
# Iterating through the json list 
for i in data['report']: 
    print(i) 
f.close() 
```

```python
import json 
import csv 
  

# Open and load json file into variable data
#================================
with open('apperian-devicedetails-20200717-modified.json') as json_file: 
    data = json.load(json_file) 

device_data = data['entries'] 
  
# prep for writing 
#================================
data_file = open('apperian-devicedetails-20200717-modified.csv', 'w') 
  

# instantiate & create the csv writer object 
#================================
csv_writer = csv.writer(data_file) 
  
# create counter for headers to csv file 
#================================
ctr = 0
  
for ddt in device_data: 
    if ctr == 0: 
  
        # Writing headers of CSV file 
        header = ddt.keys() 
        csv_writer.writerow(header) 
        ctr += 1
  
    # Writing data of CSV file 
    csv_writer.writerow(ddt.values()) 
    
data_file.close() 
```
